# Primeiro Modelo de Machine Learning com Scikit-Learn
Neste notebook, vamos criar nosso 'Hello World' usando um modelo simples de **Regressão Linear**.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

--- 
### 1. Preparando os Dados (Features e Labels)
Vamos criar um padrão matemático simples para a IA descobrir sozinho: $y = 2x + 1$

* `X`: Nossas Features (Os números de entrada)
* `y`: Nossos Labels (O resultado desejado para a entrada)

```python
X = np.array([[1], [2], [3], [4], [5]]) 
y = np.array([3, 5, 7, 9, 11])          

print("Features (X):\n", X)
print("Labels (y):", y)
```

--- 
### 2 & 3. Escolhendo, Instanciando e Treinando o Modelo
A Regressão Linear é um dos modelos mais simples de Machine Learning. 
Aqui instanciamos e em seguida mandamos ele "aprender" (`.fit`) chamando a função com os dados que criamos acima.

```python
model = LinearRegression()
model.fit(X, y)

print("Modelo treinado com sucesso!")
```

--- 
### 4. Fazendo uma previsão com um dado novo
Agora que a IA aprendeu o padrão, vamos pedir para prever o valor de $y$ quando for fornecido um novo $x$ (neste caso, $x = 6$).

Sabendo que a regra era $y = 2x + 1$, o esperado para $x=6$ seria **$13$** ($2 \times 6 + 1$).

```python
test_value = np.array([[6]])
prediction = model.predict(test_value)

print(f"Previsão calculada pela IA para x=6: {prediction[0]:.2f}")
print("Sucesso! O modelo aprendeu nosso padrão linear.")
```
